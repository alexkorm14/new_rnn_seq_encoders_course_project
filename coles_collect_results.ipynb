# Сбор метрик по обучению coles

```python
import pandas as pd
import numpy as np
import yaml
```

```python
result_path = 'drive/My Drive/pytorch-lifestream/results'
results_coles = list(filter(lambda x: 'coles' in x, os.listdir(result_path)))
results = []

for rc in results_coles:
    res = pd.read_csv(os.path.join(result_path,rc))
    config_name = rc.replace('csv','yaml').replace('metrics_','')
    with open(os.path.join('drive/My Drive/pytorch-lifestream/configs',config_name),'r') as f:
        model_config = yaml.safe_load(f)
    dataset_name = model_config['data_config'].split('.')[0]

    assert config_name in os.listdir(os.path.join('drive/My Drive/pytorch-lifestream/configs')), 'config file not found'

    res['dataset_name'] = dataset_name
    res['config_file'] = config_name

    results.append(res)

results = pd.concat(results).reset_index(drop=True).drop(columns=['num_layers','bidir'])
```

```python
for dataset_name in results['dataset_name'].unique():
    print(dataset_name)
    mask = (results['dataset_name'] == dataset_name) & (results['dataset'] == 'valid')
    display(results.loc[mask].sort_values('roc-auc',ascending=False))
    print()
```
