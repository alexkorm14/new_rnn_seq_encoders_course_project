## SberHack gender prediction

```python
data_config_name = 'sberhack_gender_prediction_data.yaml'

for model_name in ['gru','lstm','cornn','urnn','indrnn','qrnn','lmu','lem','lru','s5']:
    for hidden_size in [64,128]:
        for num_layers in [1]:
            for window in [1]:
                if model_name != 'qrnn':
                    config_name_s25 = data_config_name.replace('data',f's2t_{model_name}_{hidden_size}_{num_layers}')
                    additional_params = None

                    if model_name == 'cornn':
                        additional_params = dict(dt=5.4e-2, gamma=4.9, epsilon=4.8)
                    elif model_name == 'lmu':
                        additional_params = dict(memory_size = hidden_size * 2, theta = 100)
                    elif model_name == 'lru':
                        additional_params = dict(state_features = hidden_size // 2)

                    config_seq2target1 = dict(
                        task = {'objective': 'classification','num_classes' : 1},
                        data_config = data_config_name,
                        rnn_config = {'rnn_type' : model_name,
                                      'category_emb_dim_reduction' : 5,
                                      'hidden_state' : hidden_size,
                                      'num_layers' : num_layers,
                                      'bidir' : False,
                                      'own_head': False,
                                      'additional_params' : additional_params},
                        batch_size = 10,
                        num_workers = 2,
                      )

                    with open(os.path.join(path_to_load,config_name_s25), 'w') as f:
                        yaml.dump(config_seq2target1, f)

                    config_name_coles = data_config_name.replace('data',f'coles_{model_name}_{hidden_size}_{num_layers}')

                    config_coles = dict(
                        data_config = data_config_name,
                        rnn_config = {'rnn_type' : model_name,
                                      'category_emb_dim_reduction' : 5,
                                      'hidden_state' : hidden_size,
                                      'num_layers' : num_layers,
                                      'bidir' : False,
                                      'additional_params' : additional_params},
                        min_seq_len = 20,
                        splitter_params = dict(split_count=20,cnt_min=50,cnt_max=1000),
                        batch_size = 128,
                        num_workers = 2,
                      )

                    with open(os.path.join(path_to_load,config_name_coles), 'w') as f:
                        yaml.dump(config_coles, f)

                else:

                    config_name_s25 = data_config_name.replace('data',f's2t_{model_name}_{hidden_size}_{num_layers}_{window}')
                    additional_params = None


                    additional_params = {'window' : window}

                    config_seq2target1 = dict(
                        task = {'objective': 'classification','num_classes' : 1},
                        data_config = data_config_name,
                        rnn_config = {'rnn_type' : model_name,
                                      'category_emb_dim_reduction' : 5,
                                      'hidden_state' : hidden_size,
                                      'num_layers' : num_layers,
                                      'bidir' : False,
                                      'own_head': False,
                                      'additional_params' : additional_params},
                        batch_size = 10,
                        num_workers = 2,
                      )

                    with open(os.path.join(path_to_load,config_name_s25), 'w') as f:
                        yaml.dump(config_seq2target1, f)

                    config_name_coles = data_config_name.replace('data',f'coles_{model_name}_{hidden_size}_{num_layers}_{window}')

                    config_coles = dict(
                        data_config = data_config_name,
                        rnn_config = {'rnn_type' : model_name,
                                      'category_emb_dim_reduction' : 5,
                                      'hidden_state' : hidden_size,
                                      'num_layers' : num_layers,
                                      'bidir' : False,
                                      'additional_params' : additional_params},
                        min_seq_len = 20,
                        splitter_params = dict(split_count=20,cnt_min=50,cnt_max=1000),
                        batch_size = 128,
                        num_workers = 2,
                      )

                    with open(os.path.join(path_to_load,config_name_coles), 'w') as f:
                        yaml.dump(config_coles, f)

```
